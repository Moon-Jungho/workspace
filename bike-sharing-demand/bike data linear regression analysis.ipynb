```python
# https://www.kaggle.com/c/bike-sharing-demand/data?select=train.csv 에서 데이터를 다운받습니다.
#다운받은 파일의 압축을 해제하고 bike-sharing-demand/train.csv 데이터를 train 변수로 가져옵니다.

import pandas as pd
import numpy as np

bike = pd.read_csv('~/aiffel/bike-sharing-demand/train.csv')
df = pd.DataFrame(bike)

bike.head()
```

```python
print(df['datetime'])
```

```python
from datetime import datetime



datetime.strptime("2018-01-31", "%Y-%m-%d")
```

```python
#year, month, day, hour, minute, second 데이터 개수 시각화하기

a = pd.to_datetime(df['datetime'])
df['year'] = a.dt.year
df['month'] = a.dt.month
df['day'] = a.dt.day
df['hour'] = a.dt.hour
df['minute'] = a.dt.minute
df['second'] = a.dt.second

df.head()
```

```python
#sns.countplot 활용해서 시각화하기
import matplotlib.pyplot as plt

import seaborn as sns
sns.countplot(x="year", data=df)
plt.show()
sns.countplot(x="month", data=df)
plt.show()
sns.countplot(x="day", data=df)
plt.show()
sns.countplot(x="hour", data=df)
plt.show()
sns.countplot(x="minute", data=df)
plt.show()
sns.countplot(x="second", data=df)
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,3,1)
ax1.bar(df['year'], df['count'])
ax2 = fig.add_subplot(2,3,2)
ax2.bar(df['month'], df['count'])
ax3 = fig.add_subplot(2,3,3)
ax3.bar(df['day'], df['count'])
ax4 = fig.add_subplot(2,3,4)
ax4.bar(df['hour'], df['count'])
ax5 = fig.add_subplot(2,3,5)
ax5.bar(df['minute'], df['count'])
ax6 = fig.add_subplot(2,3,6)
ax6.bar(df['second'], df['count'])
```

```python
#X에는 문자열 데이터, 쓸모없는 데이터를 제외한 컬럼의 데이터 넣기
#count 값을 맞추고자 하므로, y 변수에 count 컬럼의 데이터 넣기
#datetime	season	holiday	workingday	weather	temp	atemp	humidity	windspeed	
#casual	registered	count	year	month	day	hour	minute	second


df_x=df[['season', 'holiday', 'workingday', 'weather', 'temp', 'humidity', 'windspeed', 'year','month']]
df_y=df['count']
X=np.array(df_x)
y=np.array(df_y)
```

```python
df_x.head()
```

```python
#X와 y 데이터를 각각 train 데이터와 test 데이터로 분리

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y)

print(X_test.shape, y_test.shape)
```

```python
#sklearn의 LinearRegression 모델 불러오기 및 학습하기

from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
#학습된 모델에 X_test를 입력해서 예측값 출력하기
predictions = model.predict(X_test)

predictions
```

```python
#모델이 예측한 값과 정답 target 간의 손실함수 값 계산하기
#mse 값과 함께 rmse 값도 계산하기

from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, predictions)
rmse = mse**(1/2)

print('mse:',mse,'rmse:',rmse)
```

```python
#x축에 X 데이터 중 temp 데이터를, y축에는 count 데이터를 넣어서 시각화하기
plt.scatter(X_test[:, 4], y_test, label="true")
plt.scatter(X_test[:, 4], predictions, label="pred")
plt.legend()
plt.show()
```

```python
#x축에 X 데이터 중 humidity 데이터를, y축에는 count 데이터를 넣어서 시각화하기

plt.scatter(X_test[:, 5], y_test, label="true")
plt.scatter(X_test[:, 5], predictions, label="pred")
plt.legend()
plt.show()

#season	holiday	workingday	weather	temp	humidity	windspeed	year	month
```
